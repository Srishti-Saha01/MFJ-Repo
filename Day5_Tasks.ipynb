```python
# Subtask 1 - Data Cleaning & Deduplication

import pandas as pd

# Load the data
PATH = "/content/raw_data_v0.1 - master.csv"

df = pd.read_csv(PATH)

# Convert 'Timestamp' column to datetime format
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

# Sort the data by 'Timestamp' in descending order
df = df.sort_values(by='Timestamp', ascending=False)

# Create a unique identifier column
df["unique_id"] = df["Verified Email"].fillna('') + df["Email Id"].fillna('') + df["WhatsApp Number"].fillna('')

# Deduplicate based on 'unique_id'
df = df.drop_duplicates(subset='unique_id', keep='first')

# Normalize 'College' and 'District' names
df["College State"] = df["College State"].str.lower().str.strip()
df["College District"] = df["College District"].str.lower().str.strip()

# Save the cleaned data
df.to_csv("cleaned_master.csv", index = False)
```

```python
# Subtask 2 - Attendance Report Generation

import pandas as pd

# Load the cleaned data
df = pd.read_csv("cleaned_master.csv")

# Convert 'Timestamp' to datetime format
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

# Define the time ranges for each day
time_ranges = {
    'Day 1': (pd.to_datetime("28 Jul 2023 18:00:00"), pd.to_datetime("29 Jul 2023 17:59:59")),
    'Day 2': (pd.to_datetime("29 Jul 2023 18:00:00"), pd.to_datetime("30 Jul 2023 17:59:59")),
    'Day 3': (pd.to_datetime("30 Jul 2023 18:00:00"), pd.to_datetime("31 Jul 2023 17:59:59")),
    'Day 4': (pd.to_datetime("31 Jul 2023 18:00:0"), pd.to_datetime("1 Aug 2023 17:59:59")),
    'Day 5': (pd.to_datetime("1 Aug 2023 18:00:00"), pd.to_datetime("2 Aug 2023 17:59:59")),
}

def check_attendance(email_id, day):
    start_time, end_time = time_ranges[day]
    attendance = df[(df['Email Id'] == email_id) & (df['Timestamp'] >= start_time) & (df['Timestamp'] <= end_time)]
    return 'Present' if not attendance.empty else 'Absent'
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned data from the CSV file
attendance_df = pd.read_csv('cleaned_master.csv')

# Calculate the attendance count for each day
attendance_count = attendance_df.iloc[:, 2:].applymap(lambda x: 1 if x == '' else 0).sum()

# Plot a bar chart
attendance_count.plot(kind='bar')
plt.xlabel('Day')
plt.ylabel('Attendance Count')
plt.title('Attendance Count by Day')
plt.show()

# Calculate the attendance percentage for each student
student_attendance = attendance_df.iloc[:, 2:].applymap(lambda x: 1 if x == '' else 0).mean(axis=1) * 100

# Plot a bar chart
plt.figure(figsize=(10,6))
student_attendance.plot(kind='bar')
plt.xlabel('Student')
plt.ylabel('Attendance Percentage')
plt.title('Attendance Percentage by Student')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Calculate the overall attendance percentage
overall_attendance = attendance_df.iloc[:, 2:].applymap(lambda x: 1 if x == '' else 0).mean().mean() * 100
print(f'Overall Attendance Percentage: {overall_attendance:.2f}%')
```

```python
# Subtask 3 - College Demographic Report

import pandas as pd
import matplotlib.pyplot as plt;

# Assuming 'data.csv' is your dataset file
df = pd.read_csv('cleaned_master.csv')

# Output Report B: College-Wise Participation Report
college_wise_report = df.groupby('College Full Name')['Email Id'].nunique().reset_index()
college_wise_report.columns = ['College Full Name', 'No. of Students']
print(college_wise_report)

# Output Report C: District-Wise Participation Report
district_wise_report = df.groupby('College District')['Email Id'].nunique().reset_index()
district_wise_report.columns = ['College District', 'Student Count']
print(district_wise_report)

# Output Report D: State-Wise Participation Report
state_wise_report = df.groupby('College State')['Email Id'].nunique().reset_index()
state_wise_report.columns = ['College State', 'No. of Students']
print(state_wise_report)

# Output Report E: Top Branches Report
branches_wise_report = df.groupby('College State')['Email Id'].nunique().reset_index()
branches_wise_report.columns = ['College State', 'No. of Students']
print(branches_wise_report)
```
